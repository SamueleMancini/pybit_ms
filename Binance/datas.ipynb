```python
# Function to fetch daily prices for a coin
def fetch_daily_prices(coin_id, start_date, end_date):
    url = f"https://api.coingecko.com/api/v3/coins/{coin_id}/market_chart/range"
    params = {
        "vs_currency": "eur",
        "from": int(start_date.timestamp()),
        "to": int(end_date.timestamp())
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        prices = {}
        for entry in data.get("prices", []):
            # Convert timestamp to date
            timestamp = int(entry[0])
            date = datetime.datetime.fromtimestamp(timestamp / 1000).strftime("%d-%m-%Y")
            prices[date] = entry[1]  # Use price as value
        return prices
    else:
        print(f"Error fetching data for {coin_id}: {response.status_code}")
        return 0

# Start and end dates
start_date = datetime.datetime(2024, 1, 1)
end_date = datetime.datetime.now(datetime.timezone.utc)

# Fetch prices for all coins
all_prices = {}
dates = set()
```

```python
COINS= [
]
```

```python
for coin in COINS:
    print(f"Fetching data for {coin}...")
    prices = fetch_daily_prices(coin, start_date, end_date)
    if not prices:
        break
    all_prices[coin] = prices
    dates.update(prices.keys())
    time.sleep(1)
```

```python
# Sort dates
sorted_dates = sorted(dates, key=lambda x: datetime.datetime.strptime(x, "%d-%m-%Y"))

# Write to CSV
output_file = "data3.csv"
with open(output_file, "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    # Write header
    header = ["Date"] + COINS
    writer.writerow(header)

    # Write rows
    for date in sorted_dates:
        row = [date] + [
            all_prices[coin].get(date, "NA") for coin in COINS
        ]  # Use "NA" if no data for that date
        writer.writerow(row)

print(f"Data saved to {output_file}")
```
